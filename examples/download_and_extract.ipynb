# Downloading and Extracting a RADAR Dataset

This notebook demonstrates how to download, verify, and extract a dataset from the [RADAR](https://www.radar-service.eu) research data repository using `radar-jupyter`.

The `download_and_extract` function:
- accepts a plain dataset ID, a RADAR ID (`RADAR/<id>`), or a DOI
- downloads the dataset TAR archive
- verifies the MD5 checksum against the RADAR metadata
- extracts the archive
- skips already-downloaded/extracted datasets on subsequent calls (with checksum re-verification)


## Setup

Install the library if not already installed (only needed once in your environment):

```python
# %pip install radar-jupyter
```

```python
from radar_jupyter import download_and_extract
```

## Download by Dataset ID

The simplest form — pass the plain RADAR dataset ID directly:

```python
dataset_id = "your-dataset-id-here"

data_path = download_and_extract(dataset_id)
print(f"Dataset available at: {data_path}")
```

## Download by DOI

You can also pass a DOI — either as a bare DOI or a full `https://doi.org/` URL:

```python
# bare DOI
data_path = download_and_extract("10.2222/your-dataset-id-here")

# or as a full DOI URL
data_path = download_and_extract("https://doi.org/10.2222/your-dataset-id-here")

print(f"Dataset available at: {data_path}")
```

## Download by RADAR ID

RADAR IDs use the `RADAR/<id>` prefix:

```python
data_path = download_and_extract("RADAR/your-dataset-id-here")
print(f"Dataset available at: {data_path}")
```

## Exploring the Extracted Files

`download_and_extract` returns a `pathlib.Path` pointing to the dataset's `data/` directory (or extraction root). You can work with the files directly:

```python
# List all files in the dataset
files = sorted(data_path.rglob("*"))
for f in files:
    print(f.relative_to(data_path))
```

## Caching Behaviour

Calling `download_and_extract` a second time with the same dataset ID skips the download and extraction. The TAR checksum is still re-verified against the RADAR API if the TAR file is still on disk:

```python
# Second call — no download, checksum re-verified, returns the cached path
data_path = download_and_extract(dataset_id)
print(f"Dataset available at: {data_path}")
```

## Configuration

Download and extraction directories, as well as the RADAR API URL, can be controlled via environment variables before importing the library:

| Variable | Default | Description |
|---|---|---|
| `RADAR_DOWNLOAD_DIR` | `downloads` | Directory where TAR archives are saved |
| `RADAR_EXTRACTION_DIR` | `extracted` | Directory where datasets are extracted |
| `RADAR_API_URL` | `https://www.radar-service.eu/radar` | RADAR API base URL |

Set them before starting the kernel, or use `os.environ` before importing `radar_jupyter`:

```python
import os
os.environ["RADAR_DOWNLOAD_DIR"] = "/data/radar/downloads"
os.environ["RADAR_EXTRACTION_DIR"] = "/data/radar/extracted"

# Import after setting env vars so the paths are picked up
from radar_jupyter import download_and_extract

data_path = download_and_extract(dataset_id)
print(f"Dataset available at: {data_path}")
```
