## Access vs. downloads per year

Draw the access-vs-download ratio as a pie chart for 2022, 2023 and 2024, side by side in one figure. Statistics are gathered per dataset from each hosting RADAR instance (and cached, so re-running is fast).


```python
import matplotlib.pyplot as plt

from radar_jupyter import plot_access_download_ratio

# Plot the access-vs-download pie chart for several years side by side.
years = [2022, 2023, 2024]
fig, axes = plt.subplots(1, len(years), figsize=(15, 5))

for year, ax in zip(years, axes):
    plot_access_download_ratio(year, ax=ax)

fig.suptitle("RADAR access vs. downloads by year", fontsize=16)
fig.tight_layout()
plt.show()
```

## Download growth across years

Sum the total downloads for each of the three years and plot them as a line chart to show how downloads ramp up over time. Reuses the cached statistics from the cell above.


```python
import matplotlib.pyplot as plt

from radar_jupyter import aggregate_access_download_stats

# Show how the total number of downloads ramps up across the three years.
# Reuses the cached per-dataset statistics gathered above, so this is fast.
years = [2022, 2023, 2024]
downloads = [aggregate_access_download_stats(year).total_downloads for year in years]

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(years, downloads, marker="o", linewidth=2, color="tab:orange")

# Annotate each point with its absolute download count.
for year, value in zip(years, downloads):
    ax.annotate(
        f"{value:,}",
        (year, value),
        textcoords="offset points",
        xytext=(0, 10),
        ha="center",
    )

ax.set_title("RADAR downloads ramp-up (2022–2024)")
ax.set_xlabel("Publication year")
ax.set_ylabel("Total downloads")
ax.set_xticks(years)
ax.margins(y=0.15)
ax.grid(True, axis="y", alpha=0.3)
fig.tight_layout()
plt.show()
```
